## Series

This section shows different ways dtypes act according to the data input in the series

```python
mydata = ['Boat', 'Car', 'Bike', 'Truck']

myseries1 = pd.Series(mydata)

print(myseries1)
```

```python
mydata = [1, 55, 99, 43]

myseries2 = pd.Series(mydata)

print(myseries2)
```

## DataFrame

```python
# Using pd.DataFrame to create a dataframe out of series data
mydfdata = [('Boat', 1), ('Car', 55), ('Bike', 99), ('Truck', 43)]

mydf = pd.DataFrame(mydfdata)

# When we call mydf you can see that there's an index column that both series are tied to
mydf

# You can think of these as spreadsheet columns and rows
```

```python
# Let's give the columns names
mydf = pd.DataFrame(mydfdata, columns=['thing', 'count'])

mydf
```

```python
# See what dtypes exist within the dataframe
mydf.dtypes
```

# Subsetting our data

```python
# To get all the columns of a dataset we can use the .columns method
df.columns
```

```python
# If we want to subset and only use a specific selectin of columns we can use this method
df = df[['title', 'description', 'publishTime',
         'duration_seconds', 'viewCount', 'likeCount',
         'commentCount']]
```

```python
# Using a certain number of rows insead
df.shape
```

```python
# Subsetting using loc
df_subset1 = df.loc[df['viewCount'] > 1_000_000]
```

```python
# Subsetting using query
df_subset2 = df.query('viewCount > 1000000')
df_subset2
```

```python
# Lets remove the NaN values, in the dataset there are certain values that does not contain video information

# Check which rows of viewCount are NaN
~df['viewCount'].isna()

# Using .loc method, we can drop the NaN rows are replace our dataframe
df = df.loc[~df['viewCount'].isna()]
df = df.loc[~df['likeCount'].isna()]
df
```

# Casting Dtypes

```python
df.dtypes
```

```python
# We want viewCount and likeCount to be an integer as they are whole numbers and do not contain NaN values anymore
df['viewCount'] = df['viewCount'].astype(int)
df['likeCount'] = df['likeCount'].astype(int)
```

```python
# publishTime is a string, we want to convert it to a datetime object
df['publishTime'] = pd.to_datetime(df['publishTime'])
```

# Creating a new column

```python
# Lets get a rate of how many likes a video gets and put it into a new column
df['like_to_view_ratio'] = df['likeCount'] / df['viewCount']
```

```python
df
```

# Adding new rows

```python
# Define the dataframe you want to add onto an existing dataframe
df_to_append = df.tail(1)
```

```python
# Using Concat method we can append it as is to the existing dataframe, check the last two rows to see the result
df_concat = pd.concat([df, df_to_append])
df_concat.tail()
```
